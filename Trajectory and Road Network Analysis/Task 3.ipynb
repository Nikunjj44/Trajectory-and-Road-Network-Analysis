### Map Matching

#### Generating UBODT (Upper-Bound Of Distance Table)
##### This basically creates a precomputed table of shortest paths for faster map matching

```python
%%bash
ubodt_gen \
  --network /content/drive/MyDrive/urban_project2/output_task1/porto_shapefile/porto_edges.shp \
  --network_id fid \
  --source u --target v \
  --delta 0.01 \
  --output /content/drive/MyDrive/urban_project2/output_task3/ubodt_wgs84.txt \
  --use_omp
```

```python
# initializing road network
road_network = Network(root_dir + "/output_task1/porto_shapefile/porto_edges.shp", "fid", "u", "v")

# converting road network into graph
road_graph = NetworkGraph(road_network)

# loading the precomputed table (from above cell)
precomputed_table = UBODT.read_ubodt_csv(root_dir + "/output_task3/ubodt_wgs84.txt")
```

```python
# creating Fast Map matcher
fast_map_matcher = FastMapMatch(road_network, road_graph, precomputed_table)

matcher_config = FastMapMatchConfig(8, 0.003, 0.0003)
```

```python
# Map Matching

match_results = []
submission = []

start_time = time.time()

for i, row in tqdm(trajectories.iterrows(),
                          total=len(trajectories),
                          desc="Matching"):

    # converting geometry to WKT string
    trajectory_wkt = row.geometry.wkt
    trip_id = row["id"]

    # map matching
    result = fast_map_matcher.match_wkt(trajectory_wkt, matcher_config)
    matched_path = list(result.opath)

    # saving detailed results
    try:
        match_results.append({
            't_number': trip_id,
            'cpath': str(list(result.cpath)),
            'opath': str(list(result.opath)),
            'offset': str([c.offset for c in result.candidates]),
            'length': str([c.length for c in result.candidates]),
            'spdist': str([c.spdist for c in result.candidates]),
            'mgeom': result.mgeom.export_wkt() if result.mgeom else ''
        })
    except Exception as e:
        match_results.append({
            't_number': trip_id,
            'cpath': '[]',
            'opath': '[]',
            'offset': '[]',
            'length': '[]',
            'spdist': '[]',
            'mgeom': ''
        })


    # to save required submission
    road_sequence = ','.join(map(str, matched_path)) if matched_path else ''
    submission.append([trip_id, road_sequence])

total_time = time.time() - start_time

print(f"Total time taken: {total_time/60:.2f} minutes")
```

```python
df_matched_results = pd.DataFrame(match_results)
df_submission = pd.DataFrame(submission, columns=['trip_id', 'road_segments'])
```
